### Function plot using matplotlib 

```python
import matplotlib.pyplot as plt 
import numpy as np
%matplotlib inline
```

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t); 
```

```python
t = np.linspace(0, 5, 50); 
```

```python
plt.figure(figsize = (8, 6)); 
plt.rcParams.update({'font.size': 16}); 
plt.plot(t, f(t), 'bo', t, f(t), 'k'); 
plt.xlabel('$t$'); 
plt.ylabel('$f(t)$'); 
plt.show(); 
```
