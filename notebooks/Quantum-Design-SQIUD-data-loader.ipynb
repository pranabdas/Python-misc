```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
plt.rcParams["figure.dpi"]=100
```

```python
def squid_load_data(filename):
    fid = open(filename, 'r')
    if fid.mode == 'r' :
        contents = fid.read()
    fid.close()

    # read each line
    contents = contents.splitlines()

    # look for the line '[Data]'
    for line in range(len(contents)):
        if (contents[line].find('[Data]') != -1):
            lineData = line
            break

    # load data columns using numpy loadtxt
    data = np.loadtxt(filename, delimiter=',', usecols=[3, 4], skiprows=lineData+2)
    return data[:, 0], data[:, 1]
```

```python
temp, moment = squid_load_data("../datafiles/CeMg3_100_MT_1kOe_SQUID_07May2012.dc.dat")
```

```python
plt.plot(temp, moment)
plt.xlabel("Temperature (K)")
plt.ylabel("Moment (emu)")
plt.show()
```
